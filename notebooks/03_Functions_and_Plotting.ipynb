# üå° What are Anomalies?

In climate and oceanography, an anomaly is:

\[ \, 	ext{anomaly} = 	ext{value} - 	ext{mean} \, \]

Used for:
- marine heatwave detection  
- climate trend analysis  
- seasonal cycle removal  
- comparing conditions to long‚Äëterm averages  

---

## üî∂ Task 3.1 ‚Äî Compute SST Anomalies

Use:

```python
sst = np.array([28.1, 28.3, 28.4, 28.0, 27.9, 28.5, 28.7, 28.6, 28.2, 28.1])
```

Steps:
1. Compute mean  
2. Compute anomalies using vectorized subtraction  
3. Print both  


````python
sst = np.array([28.1, 28.3, 28.4, 28.0, 27.9, 28.5, 28.7, 28.6, 28.2, 28.1])

mean_sst = sum(sst) / len(sst)
sst_anom = sst - mean_sst

print("Mean SST:", mean_sst)
print("SST Anomalies:", sst_anom)
````

- Functions are essential for scientific workflows because they break complex research into modular, reusable tasks that can be easily debugged and shared. This structure enables parallel execution across distributed computing resources, significantly reducing the time needed for massive data analysis. Additionally, functions facilitate reproducibility and fault tolerance, allowing scientists to precisely track data provenance and automatically restart specific failed steps without re-running the entire experiment.
- NumPy arrays are significantly better than Python lists for satellite and model data due to their homogeneous nature and contiguous memory allocation, which allow for much higher efficiency at scale. Unlike Python lists‚Äîwhich are collections of pointers to scattered objects‚ÄîNumPy arrays store raw data in a single, compact block, reducing memory usage by as much as 10 to 20 times for large scientific datasets.
- Plotting Sea Surface Temperature (SST) anomalies‚Äîthe deviation from a long-term average‚Äîis more insightful than using raw values because it filters out the dominant seasonal cycle and latitudinal gradients to reveal underlying climate signals. By removing expected background noise, anomalies allow scientists to identify unusual ocean behavior and predict extreme weather patterns, such as hurricane intensification, with much greater clarity.
- Time-series plots reveal scientific patterns by visualizing how data variables evolve over sequential intervals, making it easy to identify long-term trends, such as rising global temperatures, and cyclical patterns like seasonal vegetation shifts. They are particularly effective for detecting extreme events, as sudden spikes or dips (outliers) clearly highlight anomalies like flash floods, heatwaves, or earthquake precursors that deviate from the historical norm. By mapping these changes over time, researchers can establish baselines, determine the frequency of rare events, and identify "tipping points" where a system shifts into a new state.
